```python
#Loading the required Libraries

import tensorflow as tf

from tensorflow.keras import models, layers
```

```python
#Loading and Pre-processing the MNIST Dataset

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
#Building the Artifical Neural Network Model

artificial_neural_network_model = models.Sequential([

                                                     layers.Flatten(input_shape=(28,28)),

                                                     layers.Dense(128, activation = 'relu'),

                                                     layers.Dense(10, activation ='softmax')

                                                     ])
```

```python
#Compiling the Artificial Neural Network Model

artificial_neural_network_model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
#Training the Artificial Neural Network Model

artificial_neural_network_model.fit(x_train, y_train, epochs = 17, batch_size = 32)
```
